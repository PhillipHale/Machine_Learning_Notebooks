```python
import numpy as np
import pandas as pd

# Data Visualization
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt 
%matplotlib inline
from matplotlib import style
```

#### Use software to generate 10 observations from a normal distribution with mean, µ = 5 and standard deviation, σ = 12.

We will use the `get_truncated_normal()` function from `scipy` to instance the generator with the parameters: mean, standard deviation, and truncation range:

```python
from scipy.stats import truncnorm

def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    # https://stackoverflow.com/questions/36894191/how-to-get-a-normal-distribution-within-a-range-in-numpy
    return truncnorm((low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)


x = get_truncated_normal(mean = 5, sd = 12)

fig = plt.figure(figsize=(15,8))
ax = sns.distplot(x.rvs(10, random_state=0), kde=True)
plt.title("Histogram of normal random numbers")
plt.show()
```

With only to 10 samples, there appears to be somewhat normal. The observations appear to fit the normal density curve, while there are still some minor off setting of the data, it appears to be have more mode of overvations around 5, indicating the middle of the distrbution. 

```python
import statsmodels.api as sm
x = x.rvs(10, random_state=0)
sm.qqplot(x, fit=True, line='45')
```

There are some important deviations as it appears to curve away from the linear line. 

#### Use software to generate 100 observations from a normal distribution with mean, µ = 5 and standard deviation, σ = 12.

#### Random numbers from other distributions

```python
X1 = get_truncated_normal(mean = .5, sd = 3).rvs(100, random_state=0) # right skewed
X2 = get_truncated_normal(mean = 9, sd = 3).rvs(100, random_state=0) # left skewed,
X3 = get_truncated_normal(mean = 5, sd =.5 ).rvs(100, random_state=0) # short tailed 
X4 = get_truncated_normal(mean = 5, sd =8 ).rvs(100, random_state=0) # long tailed 

fig, ax = plt.subplots(4, sharex=True, figsize=(15,8))
sns.distplot(X1, kde=True, ax = ax[0]) # right skewed
sns.distplot(X2, kde=True, ax = ax[1]) # left skewed,
sns.distplot(X3, kde=True, ax = ax[2]) # short tailed 
sns.distplot(X4, kde=True, ax = ax[3]) # long tailed 
plt.show()
```

Next we will use the `statsmodel` api to plot a Normal Quantile

```python
import statsmodels.api as sm
fig, ax = plt.subplots(4, sharex=True, figsize=(15,8))
sm.qqplot(X1, ax=ax[0], fit=True, line='45')
sm.qqplot(X2, ax=ax[1], fit=True, line='45')
sm.qqplot(X3, ax=ax[2], fit=True, line='45')
sm.qqplot(X4, ax=ax[3], fit=True, line='45')
```
