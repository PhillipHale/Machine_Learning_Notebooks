__Objective:__ to predict profits for a food truck. Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet. The chain already has trucks in various cities and you have data for profits and populations from the cities. 


```python
import pandas as pd

# Data Visualization
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt 
%matplotlib inline
from matplotlib import style

# Statistical Packages
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# import food truck data
truck = pd.read_table("./food_truck.txt", header=None, sep=',',
                      names=['city_population', 'profits'])
truck.head()
```

## Simple Linear Regression

### Least Squares Fit
For the Advertising data, we will plot the least squares fit for the regression of `profits` onto `city_population`.

The simple linear regression formula to indicate this relationship is:
    
`profits` â‰ˆ $\beta_{0} + \beta_{1} x$ `city_population`

Next we will plot a scatter plot with a regression lint to view the distribution and see if this is a good fit for a regression problem.

```python
sns.jointplot("city_population", "profits", data=truck, kind="reg",
              color="r", height=7)
```

View 2:

```python
sns.regplot(truck.city_population, truck.profits, order = 1,
           ci = None, scatter_kws={'color':'r', 's':9})
```

### Estimating Regression Coefficient

Now we define a linear regression model with the `profits` column as the response and the `city_population` column as the feature in the `truck` dataset.

```python
# Ordinary Least Squares (ols) 

#create fitted model
lm = smf.ols(formula='profits ~ city_population', data = truck).fit()

print(lm.params,'\n')
print(lm.summary())
```
